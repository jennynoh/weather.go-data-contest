# 화재 장소 별 확인

```python
fire_fighting_loc = fire_fighting_df[['fire_firefighting.num', 'fire_firefighting.tm', 'fire_firefighting.district_1', 'fire_firefighting.district_2']]
```

```python
# location category 3별 화재건수
fire_fighting_df.groupby('fire_firefighting.location_category_3').count()
```

```python
# location category 3별 화재 피해액 평균
fire_fighting_df.groupby('fire_firefighting.location_category_3')[['fire_firefighting.property_damage']].mean()

# 평균 화재 피해금액 큰 순으로 나열 
fire_fighting_df.groupby('fire_firefighting.location_category_3')[['fire_firefighting.property_damage']].mean().sort_values('fire_firefighting.property_damage', ascending=False)
```

```python
# 피해액으로 sorting
fire_fighting_df.sort_values('fire_firefighting.property_damage', ascending=False)
```
