## 전처리 - firefighting 날짜, 시간 분리

```python
# 화재 발생 date 분리 as datetime type
fire_fighting_df['fire_date'] = fire_fighting_df['fire_firefighting.tm'].apply(lambda x: pd.to_datetime(x.split()[0]))
```

```python
# 그룹화를 위한 month 열 추출
fire_fighting_df['month'] = fire_fighting_df['fire_date'].apply(lambda x: x.month)
```

```python
# 화재발생 시간 분리 as timestamp
fire_fighting_df['fire_firefighting.tm'] = fire_fighting_df['fire_firefighting.tm'].apply(lambda x: pd.to_datetime(x))
fire_fighting_df['fire_time'] = fire_fighting_df['fire_firefighting.tm'].apply(lambda x: x.strftime('%H:%M:%S'))
```

```python
fire_fighting_df.head()
```

```python
# 필요없는 컬럼 제거
# weather_df.drop('Unnamed: 0', axis=1, inplace=True)
# datetime 객체 변환
weather_df['fire_weather.tm'] = weather_df['fire_weather.tm'].apply(lambda x: pd.to_datetime(x))
weather_df.head()
```

```python
# 필요 없는 컬럼 제거
fire_df = fire_fighting_df[['fire_firefighting.num', 'fire_firefighting.tm','fire_date','fire_time', 'fire_firefighting.district_1', 'fire_firefighting.district_2', 'fire_firefighting.fire_type_1', 'fire_firefighting.fire_type_2', 'fire_firefighting.ignition_factor_category_1', 'fire_firefighting.ignition_factor_category_2', 'fire_firefighting.casualties', 'fire_firefighting.dead', 'fire_firefighting.injury', 'fire_firefighting.property_damage', 'fire_firefighting.location_category_1', 'fire_firefighting.location_category_2', 'fire_firefighting.location_category_3']]
fire_df
```

```python
# fire_df 컬럼명 변경
fire_df.columns = ['fire_firefighting.num', 'fire_weather.tm', 'fire_date',
       'fire_time', 'fire_firefighting.district_1',
       'fire_firefighting.district_2', 'fire_firefighting.fire_type_1',
       'fire_firefighting.fire_type_2',
       'fire_firefighting.ignition_factor_category_1',
       'fire_firefighting.ignition_factor_category_2',
       'fire_firefighting.casualties', 'fire_firefighting.dead',
       'fire_firefighting.injury', 'fire_firefighting.property_damage',
       'fire_firefighting.location_category_1',
       'fire_firefighting.location_category_2',
       'fire_firefighting.location_category_3']
fire_df
```

```python
# 날씨데이터 + 화재데이터 일 기준 결합
fire_weather_df = pd.merge(weather_df, fire_df, left_on=['fire_weather.tm', 'fire_weather.district_1', 'fire_weather.district_2'] 
                           , right_on=['fire_weather.tm', 'fire_firefighting.district_1','fire_firefighting.district_2',], how='left')
fire_weather_df
```

```python
# 일자별 화재 건수 count
fire_count_df = fire_weather_df.groupby('fire_weather.tm')[['fire_firefighting.num']].count()
fire_count_df=fire_count_df.reset_index()
fire_count_df['duplicate_counts'] = fire_count_df.groupby('fire_weather.tm')['fire_weather.tm'].transform('size')
fire_count_df.head(50)
```
