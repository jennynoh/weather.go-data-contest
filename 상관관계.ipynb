# 전처리

```python
# 기상특보 datetime type 변환 
weather_special_df['fire_weather_special.start_tm'] = weather_special_df['fire_weather_special.start_tm'].apply(lambda x: pd.to_datetime(x))
weather_special_df['fire_weather_special.end_tm'] = weather_special_df['fire_weather_special.end_tm'].apply(lambda x: pd.to_datetime(x))
```

```python
# 기상특보 ~ 화재발생 matching 함수
def match_weather_special(fire_df, weather_df):
    # district_2가 일치하는 것 중에
    # 특보 시작시간 <= 화재발생시 <= 특보 종료시간
    # ...이면 해당 행의 'fire_weather_special.wrn' 값을 가져옴 

    # weather_special 빈 열 생성 
    fire_df['weather_special'] = pd.Series()
    # 지역 구분 ID 열 생성 
    fire_df['district_id'] = fire_df['fire_firefighting.district_1'] + fire_df['fire_firefighting.district_2']
    
    for i in range(len(fire_df)):        
        district = district = fire_df.iloc[i, :]['fire_firefighting.district_2']
        fire_time =  fire_time = fire_df.iloc[i, :]['fire_firefighting.tm']
        
        # 같은 지역으로 wweather_special_df 필터링
        district_weather_df = weather_df[weather_df['fire_weather_special.district_2'] == district]
        for j in range(len(district_weather_df)):
            start_time = district_weather_df.iloc[j, :]['fire_weather_special.start_tm']
            end_time = district_weather_df.iloc[j, :]['fire_weather_special.end_tm']
            if start_time <= fire_time <= end_time:
                fire_df.iloc[i, :]['weather_special'] = district_weather_df.iloc[j, :]['fire_weather_special.wrn']
                break

    return fire_df
```

```python
# 화재발생 - 기상특보 매칭
fire_weather_df = match_weather_special(fire_fighting_df, weather_special_df)
fire_weather_df
```

# 화재장소 ~ 화재발생 여부 카이제곱 상관계수
- Chi2 Statistic: 카이제곱 통계량입니다. 이 값이 클수록 두 변수 간의 관계가 더 강하게 나타납니다.
- P-value: 귀무가설이 참이라는 가정 하에 관측된 데이터에서 통계량이 나올 확률입니다. 일반적으로 p-value가 0.05 미만이면 귀무가설을 기각하고, 두 변수 간에 상관관계가 있다고 결론지을 수 있습니다.
- Degrees of Freedom: 자유도입니다. 이 값은 독립적으로 변할 수 있는 범주의 수를 나타냅니다.
- Expected Frequencies: 관찰된 빈도와 비교할 때 기대되는 빈도입니다.

# Cramer's V 상관계수

```python
import pandas as pd 
import numpy as np
from scipy.stats import chi2_contingency

# Cramer's V 함수
def cramers_v(x, y):
    # 교차표 생성 
    contingency_table = pd.crosstab(x, y)

    # 카이제곱 통계량 계산
    chi2, p, dof, expected = chi2_contingency(contingency_table)

    # 샘플사이즈
    n = contingency_table.sum().sum() 

    # Cramer's V 계산
    r, k = contingency_table.shape
    cramers_v_value = np.sqrt((chi2 / n) / min(r-1, k-1))

    return cramers_v_value
```

```python
# Cramer's V 계산
cramers_v_value = cramers_v(fire_weather_df['loc_category'], fire_weather_df['weather_special'])
print(f"Cramer's V: {cramers_v_value}")
```

# location 상위 카테고리 mapping

```python
# location 상위 카테고리 리스트 생성 
outside = ['기타', '임야']
living = ['주거']
industry = ['산업시설', '위험물/가스제조소']
transport = ['선박/항공기', '자동차/철도차량', '운수자동차시설']
gathering = ['교육시설', '생활서비스', '의료/복지시설', '집합시설', '판매/업무시설']
etc = ['기타서비스', '문화재시설']

# category 매핑 함수
def map_category(x):
    if x in outside: 
        return 'outside'
    elif x in living:
        return 'living'
    elif x in industry:
        return 'industry'
    elif x in transport:
        return 'transport'
    elif x in gathering:
        return 'gathering'
    else: 
        return 'etc'
```
