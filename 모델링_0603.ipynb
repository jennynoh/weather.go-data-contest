```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV, StratifiedKFold, cross_val_score
from sklearn.metrics import accuracy_score, recall_score, precision_score, confusion_matrix
import warnings 
warnings.filterwarnings('ignore')

# x값, y값 분리
x = df[['ta_max_t-0', 'ta_min_t-0', 'ta_max_min_t-0', 'rn_day_t-0',
       'ws_max_t-0', 'ws_ins_max_t-0', 'ws_mean_t-0', 'ws_min_t-0',
       'hm_max_t-0', 'hm_mean_t-0', 'hm_min_t-0', 'ta_max_t-1', 'ta_min_t-1',
       'ta_max_min_t-1', 'rn_day_t-1', 'ws_max_t-1', 'ws_ins_max_t-1',
       'ws_mean_t-1', 'ws_min_t-1', 'hm_max_t-1', 'hm_mean_t-1', 'hm_min_t-1',
       'ta_max_t-2', 'ta_min_t-2', 'ta_max_min_t-2', 'rn_day_t-2',
       'ws_max_t-2', 'ws_ins_max_t-2', 'ws_mean_t-2', 'ws_min_t-2',
       'hm_max_t-2', 'hm_mean_t-2', 'hm_min_t-2']]
y = df['type']

# DecisionTree Classifier 생성 
model = DecisionTreeClassifier(random_state=42)

# StratifiedKFold 
skfold = StratifiedKFold(n_splits=3)
k=0 # 교차검증 횟수
cv_acc = [] # 교차검증마다 정확도 저장 

for train_idx, test_idx in skfold.split(x, y):
    k += 1
    print(f'{k}번째 검증 데이터 셋')

    x_train, x_test = x.iloc[train_idx], x.iloc[test_idx]
    y_train, y_test = y.iloc[train_idx], y.iloc[test_idx]

    model.fit(x_train, y_train)
    y_hat = model.predict(x_test)
    # 평가지표 
    accuracy = accuracy_score(y_test, y_hat)
    precision = precision_score(y_test, y_hat, average='macro') # macro, micro, weighted 중 선택
    recall = recall_score(y_test, y_hat, average='macro')
    cm = confusion_matrix(y_test, y_hat)
    
    print(f'정확도: {accuracy:.2f}')
    print(f'정밀도: {precision:.2f}')
    print(f'재현율: {recall:.2f}')
    print(cm)
    print()
    cv_acc.append(accuracy)

print()
print(f'평균 정확도: {np.mean(cv_acc):.2f}')
```

```python
# GridSearchCV: scoring=accuracy

import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV, StratifiedKFold
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report
import warnings
warnings.filterwarnings('ignore')

x = df[['ta_max_t-0', 'ta_min_t-0', 'ta_max_min_t-0', 'rn_day_t-0',
       'ws_max_t-0', 'ws_ins_max_t-0', 'ws_mean_t-0', 'ws_min_t-0',
       'hm_max_t-0', 'hm_mean_t-0', 'hm_min_t-0', 'ta_max_t-1', 'ta_min_t-1',
       'ta_max_min_t-1', 'rn_day_t-1', 'ws_max_t-1', 'ws_ins_max_t-1',
       'ws_mean_t-1', 'ws_min_t-1', 'hm_max_t-1', 'hm_mean_t-1', 'hm_min_t-1',
       'ta_max_t-2', 'ta_min_t-2', 'ta_max_min_t-2', 'rn_day_t-2',
       'ws_max_t-2', 'ws_ins_max_t-2', 'ws_mean_t-2', 'ws_min_t-2',
       'hm_max_t-2', 'hm_mean_t-2', 'hm_min_t-2']]
y = df['type']

# 훈련 데이터와 테스트 데이터 분할
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.15, random_state=42, stratify=y)

# 의사결정 나무 모델 생성
tree_type1 = DecisionTreeClassifier(random_state=42)

# 파라미터 그리드 설정
param_grid = {
    'criterion': ['gini', 'entropy'],
    'max_depth': [None, 10, 20, 30, 40, 50],
    'min_samples_split': [2, 10, 20],
    'min_samples_leaf': [1, 5, 10]
}

# GridSearchCV 설정
cv = StratifiedKFold(n_splits=3)
grid_search = GridSearchCV(estimator=tree_type1, param_grid=param_grid, cv=cv, scoring='accuracy', n_jobs=-1)

# 모델 훈련
grid_search.fit(x_train, y_train)

# 최적의 파라미터 출력
print("Best parameters found: ", grid_search.best_params_)
print("Best cross-validation accuracy: {:.2f}".format(grid_search.best_score_))

# 테스트 데이터에 대한 예측
y_hat = grid_search.predict(x_test)

# 모델 평가
accuracy = accuracy_score(y_test, y_hat)
precision = precision_score(y_test, y_hat, average='macro') # macro, micro, weighted 중 선택
recall = recall_score(y_test, y_hat, average='macro')
cm = confusion_matrix(y_test, y_hat)

print(f'정확도: {accuracy:.2f}')
print(f'정밀도: {precision:.2f}')
print(f'재현율: {recall:.2f}')
print(cm)
print()

print("Classification Report:")
print(classification_report(y_test, y_hat))
```
