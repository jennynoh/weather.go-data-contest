# 의사결정트리
- 알고리즘
    - C4.5: 엔트로피 계수를 가장 많이 감소시키는 변수를 찾아 분리
    - CART (Classification And Regression Tree): 범주형 목표변수인 경우 지니지수를 가장 많이 감소시키는 변수로 분리
    - CHAID (Chi-Squeared Automatic Interaction Detection): p-value가 가장 작은 변수를 찾아 분리
 
- recall, precision 파라미터
    - macro: 각 클래스에 동일한 중요도 부여, 클래스 불균형 무시
    - micro: 전체 클래스의 총 합계를 사용하여 계산, 클래스의 불균형 반영
    - weighted: 각 클래스의 샘플 수를 가중치로 사용하여 계산, 클래스의 불균형 반영 

```python
# nan 값 전처리 - y 값
df['type'] = df['type'].fillna('없음')
df['ignition'] = df['ignition'].fillna('없음')
df['location'] = df['location'].fillna('없음')
df['grade'] = df['grade'].fillna('없음')

df['casualties'] = df['casualties'].fillna(-1)
df['property_damage'] = df['property_damage'].fillna(-1)
```

```python
# drop na - x 값 
df = df.dropna()
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV, StratifiedKFold, cross_val_score
from sklearn.metrics import accuracy_score, recall_score, precision_score, confusion_matrix
import warnings 
warnings.filterwarnings('ignore')

# x값, y값 분리
x = df[['ta_max_t-0', 'ta_min_t-0', 'ta_max_min_t-0', 'rn_day_t-0',
       'ws_max_t-0', 'ws_ins_max_t-0', 'ws_mean_t-0', 'ws_min_t-0',
       'hm_max_t-0', 'hm_mean_t-0', 'hm_min_t-0', 'ta_max_t-1', 'ta_min_t-1',
       'ta_max_min_t-1', 'rn_day_t-1', 'ws_max_t-1', 'ws_ins_max_t-1',
       'ws_mean_t-1', 'ws_min_t-1', 'hm_max_t-1', 'hm_mean_t-1', 'hm_min_t-1',
       'ta_max_t-2', 'ta_min_t-2', 'ta_max_min_t-2', 'rn_day_t-2',
       'ws_max_t-2', 'ws_ins_max_t-2', 'ws_mean_t-2', 'ws_min_t-2',
       'hm_max_t-2', 'hm_mean_t-2', 'hm_min_t-2']]
y = df['type']

# DecisionTree Classifier 생성 
model = DecisionTreeClassifier(random_state=42)

# StratifiedKFold 
skfold = StratifiedKFold(n_splits=3)
k=0 # 교차검증 횟수
cv_acc = [] # 교차검증마다 정확도 저장 

for train_idx, test_idx in skfold.split(x, y):
    k += 1
    print(f'{k}번째 검증 데이터 셋')

    x_train, x_test = x.iloc[train_idx], x.iloc[test_idx]
    y_train, y_test = y.iloc[train_idx], y.iloc[test_idx]

    model.fit(x_train, y_train)
    y_hat = model.predict(x_test)
    # 평가지표 
    accuracy = accuracy_score(y_test, y_hat)
    precision = precision_score(y_test, y_hat, average='macro') # macro, micro, weighted 중 선택
    recall = recall_score(y_test, y_hat, average='macro')
    cm = confusion_matrix(y_test, y_hat)
    
    print(f'정확도: {acc:.2f}')
    print(f'정밀도: {precision:.2f}')
    print(f'재현율: {recall:.2f}')
    print(cm)
    print()
    cv_acc.append(acc)

print()
print(f'평균 정확도: {np.mean(cv_acc):.2f}')
```
